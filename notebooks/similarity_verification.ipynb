`similarities.py`의 현재 구현(`compute_similarity`)을 그대로 사용하여 5×6 토이 데이터(User×Item)에 대해 모든 유사도 행렬을 계산한다.

```python
import sys, warnings, importlib
import numpy as np
import pandas as pd

sys.path.insert(0, '..')          # imputation_project 루트
warnings.filterwarnings('ignore')

import utils.similarities as sim_mod
importlib.reload(sim_mod)          # 최신 파일 반영

from utils.similarities import compute_similarity, available_methods

print("사용 가능한 유사도 메서드:", available_methods())
```

```python
# 토이 데이터 (User × Item, 5명 × 6아이템)
data = [
    [3, np.nan, 4, np.nan, np.nan, np.nan],
    [np.nan, 5, np.nan, 3, 5, 3],
    [3, 4, 4, 3, 4, 3],
    [1, 2, 1, np.nan, 2, np.nan],
    [1, np.nan, 5, np.nan, np.nan, np.nan],
]

USER_LABELS = ['U1', 'U2', 'U3', 'U4', 'U5']
ITEM_LABELS = ['I1', 'I2', 'I3', 'I4', 'I5', 'I6']

train_df = pd.DataFrame(data, index=USER_LABELS, columns=ITEM_LABELS)
print("=== 토이 데이터 (User × Item) ===")
display(train_df)
```

```python
# similarities.py의 compute_similarity로 모든 유사도 행렬 계산 및 출력
METHODS = available_methods()

for method in METHODS:
    try:
        S = compute_similarity(train_df, method)
        df_S = pd.DataFrame(S, index=USER_LABELS, columns=USER_LABELS).round(4)
        print(f"\n{'='*50}")
        print(f"  {method.upper()}")
        print(f"{'='*50}")
        display(df_S)
    except Exception as e:
        print(f"\n[{method}] ERROR: {e}")
```
